### Explore with WordCloud

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# Edit stop words to focus more on nuances and take away the noise
stopwords = set(STOPWORDS)
stopwords.add('rt')
stopwords.add('amp')
```

```python
up_wc = WordCloud(width=1600, height=800, background_color='white', relative_scaling=1, stopwords=stopwords, colormap='winter').generate(' '.join(i for i in combined_df['cleaned_text'][combined_df[0]=='up']))

plt.figure(figsize=(20,10))
plt.imshow(up_wc)
plt.axis("off")
plt.show()
```

```python
down_wc = WordCloud(width=1600, height=800, background_color='white', relative_scaling=1, stopwords=stopwords, colormap='autumn').generate(' '.join(i for i in combined_df['cleaned_text'][combined_df[0]=='down']))

plt.figure(figsize=(20,10))
plt.imshow(down_wc)
plt.axis("off")
plt.show()
```
