```python
from sklearn.decomposition import PCA
```

**PCA**

```python
PCA = PCA(n_components = 50, random_state=42)
PCA_tfidf = PCA.fit_transform(X_tfidf.todense())
```

```python
plt.scatter(PCA_tfidf.T[0], PCA_tfidf.T[1], c=y_le, cmap='magma')
plt.colorbar()
```
