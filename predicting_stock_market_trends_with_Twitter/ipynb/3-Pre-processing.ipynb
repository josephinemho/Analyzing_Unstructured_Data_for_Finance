### Clean data:

```python
# Getting ALL data out of a collection
tweets_list = [document for document in completed_collection.find()]
tweets_df = pd.DataFrame(tweets_list)
tweets_df.head()
```

```python
# Make text into a list (instead of series) so we can apply functions to it
tweets_text = []
for i in range(87289):
    tweets_text.append(tweets_list[i]['text'])
```

```python
type(tweets_text)
```

```python
import re
```

```python
def clean_text(text):
    text = re.sub('(http\S+)', '', text)
    text = re.sub('(http(s)?:\/\/.)?(www\.)?[-a-zA-Z0-9@:%._\+~#=]{2,256}\.[a-z]{2,6}\b([-a-zA-Z0-9@:%_\+.~#?&//=]*)', '', text)
    text = re.sub('[\W]', ' ', text)
    text = re.sub('(\@.*?(?=\s))', ' ', text)
    text = re.sub(, ' ', text)
    
    text = re.sub('([\s]{2,})', ' ', text)
    text = text.lower()
    return text
```

```python
cleaned_text = []
for t in tweets_text:
    cleaned_text.append(clean_text(t))
```

```python
tweets_df['cleaned_text'] = cleaned_text
```
