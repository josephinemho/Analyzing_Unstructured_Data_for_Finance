```python
start = dt.datetime(2008,1,1)
end = dt.datetime(2017,6,1)
AAPL_df = web.DataReader('AAPL', 'google', start, end)
```

```python
AAPL_df['Diff'] = AAPL_df['Close'].diff()
```

```python
AAPL_df
```

```python
def make_binary(data):
    data_list = []
    for d in data:
        if d < 0:
            data_list.append(-1)
        elif d > 0:
            data_list.append(1)
        elif d == 0:
            data_list.append(0)
        else:
            data_list.append(None)
    return data_list
```

```python
AAPL_df['Diff_Encoded'] = make_binary(AAPL_df['Diff'])
```
