### Clean data:

```python
# Getting ALL data out of a collection
tweets_list = [document for document in completed_collection.find()]
tweets_df = pd.DataFrame(tweets_list)
tweets_df.head()
```

```python
tweets_df['Date'] = tweets_df['timestamp'].dt.date
```

```python
tweets_df
```

```python
# Make text into a list (instead of series) so we can apply functions to it
tweets_text = []
for i in range(87289):
    tweets_text.append(tweets_list[i]['text'])
```

```python
type(tweets_text)
```

```python
import re
```

```python
def clean_text(text):
    text = re.sub('(http\S+)', '', text)
    text = re.sub('(http(s)?:\/\/.)?(www\.)?[-a-zA-Z0-9@:%._\+~#=]{2,256}\.[a-z]{2,6}\b([-a-zA-Z0-9@:%_\+.~#?&//=]*)', '', text)
    text = re.sub('[\W]', ' ', text)
    text = re.sub('(\@.*?(?=\s))', ' ', text)
    text = re.sub('([\s]{2,})', ' ', text)
    return text
```

```python
cleaned_text = []
for t in tweets_text:
    cleaned_text.append(clean_text(t))
```

```python
cleaned_text[0]
```

```python
tweets_df['cleaned_text'] = cleaned_text
```

### Vectorize data (fit & transform on all tweets):

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
start = datetime.now()

tfidf = TfidfVectorizer(ngram_range=(1,3), min_df=2)
X_tfidf = tfidf.fit_transform(tweets_df['cleaned_text'])

end = datetime.now()
print(end - start)
```
