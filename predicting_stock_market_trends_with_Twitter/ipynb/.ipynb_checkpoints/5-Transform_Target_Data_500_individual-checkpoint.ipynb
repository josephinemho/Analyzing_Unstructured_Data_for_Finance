```python
def get_change_in_Close(df):
    list_of_df = []
    for s in df['Stock'].unique():
        try_df = df.loc[df['Stock']==s]
        try_df = try_df.sort(columns='Date')
        try_df['Diff'] = try_df['Close'].diff()
        list_of_df.append(try_df)
        
    return list_of_df
```

```python
list_of_df_diff = get_change_in_Close(stocks_df)
```

```python
tweets_text = []
for i in range(87289):
    tweets_text.append(tweets_list[i]['text'])
```

```python
import re
```

```python
def clean_text(text):
    text = re.sub('(http(s)?:\/\/.)?(www\.)?[-a-zA-Z0-9@:%._\+~#=]{2,256}\.[a-z]{2,6}\b([-a-zA-Z0-9@:%_\+.~#?&//=]*)', '', text)
    text = re.sub('[\W]', ' ', text)
    return text
```

```python
type(tweets_text)
```

```python
clean_text(tweets_text)
```

```python
tweets_df['cleaned_text'] = tweets_text
```
